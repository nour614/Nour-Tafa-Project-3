# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df2000 = pd.read_csv("Data/final_tmdbdata2000.csv.gz")
df2001 = pd.read_csv("Data/final_tmdbdata2001.csv.gz")

df2001.info()
```

```python
df2000.info()
```

```python
# This line of code concatenates the 2 dataframes 
frames = [df2000, df2001]

df = pd.concat(frames,axis=0,ignore_index = True)
```

```python
df.info() # This line of code allows me to check the data
```

How many movies had valid financial information (values > 0 for budget OR revenue)?

```python
earnings = df[df["budget"]>0]
earnings.head()
```

```python
len(earnings) # This line of code tells me how many movies had positive financial 
```

```python
revenue = df[df["revenue"]>0]
revenue
```

```python
len(revenue)
```

```python
# to find out which movies have both budget and revenue above 0,without duplicates, we use this.
budget = df["budget"] > 0
revenue = df["revenue"] > 0
```

```python
combine = budget | revenue
combine
```

```python
combine_df = df[combine]
combine_df
```

```python
len(combine_df) # this tells how many movies fit both budget and revenue
print(f"There were {len(combine_df)} movies with financial information")
```

```python
sns.histplot(data=combine_df, x="certification");
```

How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
combine_df["certification"].value_counts()
```

```python
rated_g = combine_df[combine_df['certification'] == 'G']
rated_pg = combine_df[combine_df['certification'] == 'PG']
rated_pg13 = combine_df[combine_df['certification'] == 'PG-13']
rated_r = combine_df[combine_df['certification'] == 'R']
rated_nr = combine_df[combine_df['certification'] == 'NR']
```

```python
print(f'There are {len(rated_pg)} rated PG movies')
print('')
print(f'There are {len(rated_r)} rated R movies')
print('')
print(f'There are {len(rated_g)} rated G movies')
print('')
print(f'There are {len(rated_pg13)} rated PG-13 movies')
print('')
print(f'There are {len(rated_nr)} rated NR movies')
```

What is the average revenue per certification category?

```python

# set display options
import locale
locale.setlocale(locale.LC_ALL, '') # to put commas in large numbers

# use groupby to get revenue for each certification category
revenue_per_certification = df.groupby('certification')['revenue'].mean()

# loop through series to print
for index, value in revenue_per_certification.items():
    
    # restrict to categories asked for
    if index in ['G', 'PG', 'PG-13', 'R']:
        
        # format numbers: 2 decimal places and commas
        formatted_value = locale.format_string('%.2f', value, grouping = True)
        print(f"The average revenue of {index}-rated movies is ${formatted_value}.")
```

What is the average budget per certification category?

```python
# use groupby to get budget for each certification category
budget_per_certification = df.groupby('certification')['budget'].mean()

# loop through series to print
for index, value in budget_per_certification.items():
    
    # restrict to categories asked for
    if index in ['G', 'PG', 'PG-13', 'R']:
        
        # format numbers: 2 decimal places and commas
        formatted_value = locale.format_string('%.2f', value, grouping = True)
        print(f"The average budget of {index}-rated movies is ${formatted_value}.")
```
